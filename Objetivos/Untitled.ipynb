## UBICTUM | Visualización de datos RRRP
### RECIBO, ACOMODO, SURTIDO

```r
data <- read.csv(file = "base_rpv.csv")
```

```r
 unique(data$Proceso)
```

```r
id_recibo <- "Contabilización de entrada de mercancías"
```

```r
rb <- data[ data$Proceso == id_recibo , ]
```

# RECIBO

```r
library(repr)
options(repr.plot.width=18, repr.plot.height=10, res = 2000)
```

```r
par(mfrow = c(1,2))
plot(rb$Renglones, rb$Volumen.m3, lwd = 7, col = "#00968850",
    main = "Renglones vs Volumen")
grid()


plot(rb$Renglones.por.hora, rb$Volumen.m3.por.hora, lwd = 7, col = "#303F9F50",
    main = "Renglones vs Volumen por hora")
grid()
```

```r
options(repr.plot.width=20, repr.plot.height=15)
```

```r
par(mfrow = c(2,2))

## Renglones
hist(rb$Renglones, prob = T, col = "#F57F1780")
lines(density(rb$Renglones), lwd = 5, col = "#303F9F70")
grid()

## Piezas
hist(rb$Piezas, prob = T, col = "#9FA8DA70")
lines(density(rb$Piezas), lwd = 5, col = "#303F9F")
grid()

## Volumen
hist(rb$Volumen.m3, prob = T, col = "#EF535070", main = "Histograma de Volumen recibido por usuario en un turno")
lines(density(rb$Volumen.m3), lwd = 5, col = "#303F9F90")
grid()

## Tiempo
hist(rb$Tiempo, prob = T, col = "#FFF59D60", main = "Histograma de tiempo en sistema por usuario")
lines(density(rb$Tiempo), lwd = 5, col = "red")
grid()
```

```r
par(mfrow=c(2,2))

plot(rb$Piezas.por.hora, rb$Renglones.por.hora)
grid()
```
